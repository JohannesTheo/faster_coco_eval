```python
import json
import logging
from faster_coco_eval.curves import match, plot_curve

logging.root.setLevel("INFO")
logging.debug("Запись.")
```

```python
def load(file):
    with open(file) as io:
        _data = json.load(io)
        
    return _data
```

```python
prepared_coco_in_dict = load('../tests/data/eval_all_coco.json')
prepared_anns         = load('../tests/data/result_annotations.json')
```

```python
match_results = ['match_1', match(prepared_anns, prepared_coco_in_dict, category=None)]
```

```python
threshold_iou = 0.5

plot_curve([match_results], threshold_iou, label_name='test')
```
